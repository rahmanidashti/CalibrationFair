```python
import pandas as pd
```

```python
ratings_data = pd.read_json("raw/renttherunway_final_data.json", lines=True)
```

```python
ratings_data.head()
```

```python
# convert user and item ids to user and item index (from 0 to n: the number of users/items)

uid_to_uidx = dict()
iid_to_iidx = dict()

# user and item index counter
uidx_cnt = 0
iidx_cnt = 0

for eachsample in ratings_data.itertuples(index=True):
    # users index
    if eachsample.user_id not in uid_to_uidx.keys():
        uid_to_uidx[eachsample.user_id] = uidx_cnt
        uidx_cnt += 1

    # items index
    if eachsample.item_id not in iid_to_iidx.keys():
        iid_to_iidx[eachsample.item_id] = iidx_cnt
        iidx_cnt += 1
```

```python
# to write the rating data using new indices
import math
dataset_data = open("ClothingFit_data.txt", 'w')

for eachsample in ratings_data.itertuples(index=True):
    try:
        if not math.isnan(eachsample.rating):
            dataset_data.write(f"{str(uid_to_uidx[eachsample.user_id])},{str(iid_to_iidx[eachsample.item_id])},{str(int(eachsample.rating))}\n")
    except:
        print(eachsample.rating)

dataset_data.close()
```

```python
from collections import defaultdict

category_data = defaultdict(set)

for eachsample in ratings_data.itertuples(index=True):
    if eachsample.item_id in category_data.keys():
        category_data[eachsample.item_id].add(eachsample.category)
    else:
        category_data[eachsample.item_id] = {eachsample.category}
```

```python
# to write the movie data using new indices

genre_data = open("ClothingFit_genre.txt", 'w')

for eachsample in ratings_data.itertuples(index=True):
    category_genre = ""
    if eachsample.item_id in iid_to_iidx.keys():
        categories = "|".join(category_data[eachsample.item_id])
        genre_data.write(f"{str(iid_to_iidx[eachsample.item_id])},{categories}\n")

genre_data.close()
```
